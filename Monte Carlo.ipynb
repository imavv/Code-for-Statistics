```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mc_approx_pi(n_points, seed=None, plot=False):
    
    # Generate random numbers
    np.random.seed(seed)
    data = np.random.random(size=(n_points,2))
    
    # Masking
    in_circle = np.where(data[:,0]**2 + data[:,1]**2 <= 1, True, False)
    data_in_circle = data[in_circle,:]
    data_out_circle = data[~in_circle,:]
    
    # Store results
    dots_in_circle = len(data_in_circle)
    approx_result = 4*dots_in_circle/len(data)
    
    # Plot
    if plot:
        plt.figure(figsize=(5,5))

        plt.scatter(data_in_circle[:,0], data_in_circle[:,1], color='tab:orange',s=3)
        plt.scatter(data_out_circle[:,0], data_out_circle[:,1], color='tab:blue' ,s=3)
        
        print(f'Pi={approx_result} given {n_points} data points')
        plt.show()
    
    return approx_result
```

```python
x = mc_approx_pi(10000, seed=42, plot=True)
```

```python
data_points = np.arange(10,10000,10)
approx = [mc_approx_pi(n,seed=42) for n in data_points]
actual_pi = [np.pi for _ in data_points]
```

```python
# Compare
plt.figure(figsize=(15,7))

plt.plot(data_points, approx)
plt.plot(data_points, actual_pi)
plt.grid()

plt.show()
```
